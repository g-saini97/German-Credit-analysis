```python
# Importing the necessary libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_validate
from sklearn.linear_model import SGDClassifier

```

```python
numerical_values = ['Credit amount','Age','Duration']
categorical_values = ['Sex','Job','Housing','Saving accounts','Checking account','Purpose']
```

```python
##  Filling in the NAN/missing values in Saving accounts and Checking accounts.

for cat in categorical_values:
    df[cat] = df[cat].fillna(df[cat].mode().values[0])
```

```python
# data without any null values

df.head()
```

```python
## ONE HOT ENCODING

# generate binary values using get_dummies
dum_df = pd.get_dummies(df, columns=['Sex','Housing','Saving accounts','Checking account','Purpose'] )

# merge with main df
f_df = df.merge(dum_df)
f_df.drop(columns= categorical_values)

### Not going to One hot encode the'RISK' feature as it will serve as the target for the classification precess in the
### end. SO we will be Lable Encoding it in the end instead.
```

### Applying Log Transformation to fix the skew.

```python
## Applying Log transformation to the numerical values.
num_log = np.log(df[['Age', 'Credit amount','Duration']])
f_df[['Age', 'Credit amount','Duration']] = np.log(f_df[['Age', 'Credit amount','Duration']]) # also log transforming the 
                                                                                              # data that will be used in the
                                                                                              # classification process.  
```

```python
# Histogram after correcting for skewness
plt.figure(figsize = (8, 5))
x = num_log["Age"]
  
plt.hist(x, bins = 15, edgecolor = "black",)
plt.title("Histogram According to Age ")
plt.xlabel("Age")
plt.ylabel("count")
```

```python
# Histogram after correcting for skewness
plt.figure(figsize = (8, 5))
x = num_log["Credit amount"]
  
plt.hist(x, bins = 15, edgecolor = "black",)
plt.title("Histogram According to Credit Amount")
plt.xlabel("Credit amount")
plt.ylabel("count")
```

```python
# Histogram after correcting for skewness
plt.figure(figsize = (8, 5))
x = num_log["Duration"]
  
plt.hist(x, bins = 15, edgecolor = "black",)
plt.title("Histogram According to Duration ")
plt.xlabel("Duration")
plt.ylabel("count")
```

```python
## Feature scaling the numerical features for the modeling process
feature_scaler = StandardScaler()
data_scaled = feature_scaler.fit_transform(num_log)
```

```python
## Elbow Method to find the optimal K value
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    model = KMeans(n_clusters=k)
    model = model.fit(data_scaled)
    Sum_of_squared_distances.append(model.inertia_)
plt.figure(figsize=(15,9))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method to find the optimal k value')
plt.show()
```

```python
# Using K = 3 as suggested by the Elbow curve.

kmeans = KMeans(n_clusters = 3, random_state = 0)
kmeans.fit(data_scaled)
predict = kmeans.predict(data_scaled)

centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Using PCA to visualize.

pca = PCA(n_components = 3)
principalComponents = pca.fit_transform (data_scaled)
principalDf = pd.DataFrame(data = principalComponents, columns=['PC1','PC2','PC3']) 
principalDf.head()
```

```python
x = principalComponents[:,0]
y = principalComponents[:,1]
z = principalComponents[:,2]
```

```python
fig = plt.figure(figsize=(50, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z,
           linewidths=1, alpha=.9,
           edgecolor='k',
           s = 100,
           c=predict)
plt.show()
```

```python
# Lable encoding Risk for classification.

lable_encoder = LabelEncoder()
f_df["Risk"] = lable_encoder.fit_transform(df["Risk"])
```

```python
f_df.head()
```

```python
# Separating features from targets.

X = f_df.drop(columns= ['Risk','Sex','Job','Housing','Saving accounts','Checking account','Purpose']) # Only keeping the 
                                                                                                      # numerical and OHE 
                                                                                                      # features
Y = f_df['Risk']
```

```python
X
```

```python
# Splitting the Test and Train Sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# initiating the classifier of my choosing

clf =  SGDClassifier()
```

```python
# Initiating the Cross validation

cv_results = cross_validate(clf , X_train, Y_train, cv = 8)
```

```python
# Results of the cross validation

print("Results of the cross validation are : ",cv_results['test_score'])
```

```python
print("Mean of the Scores",cv_results['test_score'].mean(),"\n\nStandard Deviation of the scores : " ,cv_results['test_score'].std() )
```
